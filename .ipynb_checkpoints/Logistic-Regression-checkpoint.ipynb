```python
## importing important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline

warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_classification
```

```python
#create the dataset
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42)
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
# implement logistic regression
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# score=accuracy_score(y_test,y_pred)
# print(score)
# cm=confusion_matrix(y_test,y_pred)
# print(cm)
# print=classification_report(y_test,y_pred)

print(accuracy_score(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

### Hyperparameter Tuning and Cross validation

```python
model=LogisticRegression()
# Hyperparamteter tuning using different parameters
penalty=["l1", "l2", "elasticnet"]
c_values=[100,10,1,0,0.1,0.01]
solver=["newton-cg","lbfgs", "liblinear", "sag","saga"]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
#GridSearchCV to find the best parameter for the dataset
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,
                  scoring="accuracy",cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

### Randomized SearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
randomcv=RandomizedSearchCV(estimator=model,
                            param_distributions=params,cv=5,scoring="accuracy")
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

#### Logistic regression for multiclassification problem

```python
#create the dataset
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=42)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class="ovr")
logistic.fit(X_train,y_train)
y_pred=logistic.predict(X_test)
```

```python
y_pred
```

```python
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

### Logistic regression for Imbalanced Dataset

```python
# Create and visualize a synthetic imbalanced classification dataset.
from collections import Counter
from sklearn.datasets import make_classification
```

```python
# imbalanced classification dataset
X, y = make_classification(
    n_samples=10000,         
    n_features=2,             
    n_clusters_per_class=1,   
    n_redundant=0,            
    weights=[0.99],     # Imbalance ratio: 99% of one class, 1% of the other
    random_state=10           
)
```

```python
X
```

```python
Counter(y)
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])

# Plot the data
sns.scatterplot(x='Feature 1', y='Feature 2', hue=y, data=df, palette='viridis', alpha=0.5, edgecolor=None)
plt.title("Imbalanced Classification Dataset")
plt.show()
```

```python
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
class_weight
```

```python
model=LogisticRegression()
# Hyperparamteter tuning using different parameters
penalty=["l1", "l2", "elasticnet"]
c_values=[100,10,1,0,0.1,0.01]
solver=["newton-cg","lbfgs", "liblinear", "sag","saga"]
class_weight=[{0:w,1:y} for w in [1,10,50,100] for y in [1,10,50,100]]
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver,class_weight=class_weight)
```

```python
params
```

```python
from sklearn.model_selection import StratifiedKFold
#GridSearchCV to find the best parameter for the dataset
from sklearn.model_selection import GridSearchCV

cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=params,
                  scoring="accuracy",cv=cv,n_jobs=-1)
```
